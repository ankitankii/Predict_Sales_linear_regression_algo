```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
masterframe_men = pd.read_csv('final_men_file.csv') #final file (combined train and test file of MenClothing) for ML work
#print(masterframe_men)

master_frame_test



iv=masterframe_men.iloc[0:72,[4,9,10,8]]   #working on specific columns CPI,Earnings or wages  in dollars per hour,
                                          #Exports,Finance Rate on Personal Loans at Commercial Banks, 24 Month Loan
dv=masterframe_men.iloc[0:72,[18]]



x_train, x_test, y_train, y_test = train_test_split (iv, dv, test_size = 0.35, random_state = 0)


 

```

```python
from sklearn.linear_model import LinearRegression
lnr = LinearRegression()
lnr.fit (x_train, y_train ) 

predicted_y = lnr.predict(x_test)
print("predicted y records:")
#print(predicted_y)

predicted_y = predicted_y.reshape(-1,1)
print(predicted_y)
#predicted_y.to_csv('predicted_sales_for_menclothing.csv')
np.savetxt('predicted_sales_for_menclothing.csv', predicted_y, delimiter=',',fmt='%5s') 

```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from math import sqrt
print("RMSE is:-")
print(np.sqrt(metrics.mean_squared_error(y_test, predicted_y)))

```
