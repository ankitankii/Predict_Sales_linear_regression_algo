```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
accident=pd.read_csv('C:/python/road_acc_pr/Accidents_2016.csv')
casuality=pd.read_csv('C:/python/road_acc_pr/Casuality_details.csv')
vehicle=pd.read_csv('C:/python/road_acc_pr/Vehicle_details.csv')
makemodel=pd.read_csv('C:/python/road_acc_pr/MakeModel2016.csv')
accident.replace(0,np.nan)
accident.dropna().head()
casuality.dropna().head()
vehicle.dropna().head()
```

```python
merged_acc_veh = pd.merge(accident,vehicle, on= "Accident_Index", how= "inner")
merged_acc_veh
merged_acc_veh_cas = pd.merge(merged_acc_veh,casuality, on= "Accident_Index", how= "inner")
#merged_acc_veh_cas.to_csv('C:/python/road_acc_pr/merged_acc_veh_cas.csv')
merged_acv=pd.read_csv('C:/python/road_acc_pr/merged_acc_veh_cas.csv')
merged_acv
```
